```python
from tkinter import filedialog
from tkinter import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import scipy
from scipy.cluster import hierarchy
import matplotlib.cm as cm
import tkinter as tk
def browse():
    global file_path
    filename = filedialog.askopenfilename()
    file_path.set(filename)
    p = e1.get()
    dataset=pd.read_csv(p)
    names=list(dataset.columns.values)
    parent = Listbox(master,height=15)
    for name in names:
        parent.insert(END, name)
    parent.place(x=200, y=275)
    
def comp():
    
    global a
    p = e1.get()
    if p == "":
        Label(master,text="*required",fg='red',font=("Times", "10")).place(x=319, y=172)
    a = e2.get()

    if a.isalpha() == True:
        Label(master,text="*required and must be number",fg='red',font=("Times", "10")).place(x=205, y=231)
    if a.isdigit() == True:
        q=a
        #top=Toplevel()
        #top.geometry('639x359')
        #top.title("RESULTS")
        dataset=pd.read_csv(p)
        print ("Data contents")
        print (dataset.head())
        #Label(top,text="Data Contents",fg='red',font=("Times", "19")).place(x=30,y=219)
        #Label(top,text= dataset.head(),font=("Times","15")).place(x=195,y=255)
        X=dataset.values
        plt.title("DENDOGRAM")
        dendro=hierarchy.dendrogram(hierarchy.linkage(X,method='ward'))
        from sklearn.cluster import KMeans
        kmeans_1=KMeans(n_clusters=int(q))
        kmeans_1.fit(X)
        cluster_pred=kmeans_1.predict(X)
        cluster_pred_2=kmeans_1.labels_
        cluster_center=kmeans_1.cluster_centers_
        print("clusters co-ordinates :")
        print(cluster_center)
        #Label(top,text="clusters co-ordinates : ",fg='red',font=("Times","19")).place(x=30, y=15)
        #Label(top,text=cluster_center,font=("Times","15")).place(x=145, y=55)
        plt.figure(figsize=(5,3))
        x = np.arange(10)
        ys = [i+x+(i*x)**2 for i in range(10)]
        colors = cm.rainbow(np.linspace(0, 1, len(ys)))
        #for w in len(dataset.columns):
        for i in range(0,int(q)):
            plt.scatter(X[cluster_pred==i,0],X[cluster_pred==i,1], s = 100, c=colors[i], label ="cluster %s"%(i) )
        plt.scatter(cluster_center[:,0],cluster_center[:,1], s = 300, c = 'yellow', label = 'Centroids')
        plt.title('KMeans-Clustering ')
        plt.legend()
        plt.show()

master = Tk()
master.title("DATA MINING")
master.geometry('409x599')
master.config(bg='gold3')
global q
Label(master, text="Input file \n",bg='gold3',font=("Times", "14", "bold italic")).place(x=34,y=147)
Label(master,text="Number of cluster",bg='gold3',font=("Times", "14", "bold italic")).place(x=34, y=195)
Label(master,text="First row headers",bg='gold3',font=("Times", "14", "bold italic")).place(x=34, y=255)

    
file_path = StringVar()
e1 = Entry(master,textvariable=file_path,width=31)
e1.place(x=120,y=152)


b1= Button(master,text='Browse',command=browse).place(x=319,y=152)
e2 = Entry(master,width=27)
e2.place(x=205, y=199)

Label(master, text="CLUSTERING",bg='gold3' ,font=("Times new roman Bold",21)).place(x=120,y=44)



b4 = Button(master, text='Submit', command=comp).place(x=325,y=507)
b5= Button(master, text='quit',command=master.destroy).place(x=329,y=539)
mainloop( )
```
